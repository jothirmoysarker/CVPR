### Import TensorFlow and Other libraries

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist

import os
import random
import numpy as np
import matplotlib.pyplot as plt
```

### Load the dataset and display the shapes of training and test data

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print()
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

### Import Matplotlib for data visualization

```python
x_train = x_train.reshape(-1, 28*28).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28*28).astype("float32") / 255.0
```

### Flatten the 28×28 images into vectors and normalize pixel values to [0, 1]

```python
model = keras.Sequential([
    keras.Input(shape=(28*28,)),
    layers.Dense(units=256, activation='relu'),
    layers.Dense(units=128, activation='relu'),
    layers.Dense(units=10, activation='softmax')
])
model.summary()
```

### Define a fully connected neural network model for digit classification

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

### Train the model on the training data and validate using a hold-out split

```python
h = model.fit(x=x_train, y=y_train, epochs=5, batch_size=64, validation_split=0.3)
```

Save the Model

```python
model.save("NN model.keras")
```

Evaluate the Model

```python
load_model = tf.keras.models.load_model("mnist.keras")
load_model.evaluate(x_test, y_test)
```
